```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x):
    return 3*x**2 - 4*x + 5
```

```python
f(6)
```

```python
xs = np.arange(-5, 5, 0.25)
ys = f(xs)
plt.plot(xs , ys)
plt.show()
```

```python
h = 0.00001
x = 2/3
(f(x + h) - f(x))/h
```

```python
a = 2.0
b = -3.0
c = 10.0
d = a*b + c
print(d)
```

```python
h = 0.0001

a = 2.0
b = -3.0
c = 10.0

d1 = a * b + c
a += h
d2 = a * b + c
print('d1', d1)
print('d2', d2)
print('slope', (d2 - d1)/h)
```
